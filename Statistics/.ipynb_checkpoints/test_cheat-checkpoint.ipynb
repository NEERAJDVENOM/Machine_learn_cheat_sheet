# One sample z test
### Conditions for Z test
    1. Random samples.
    2. Each observation should be independent of each other
        - Sampling with replacement
        - If sampling without replacement, the sample size should not be more than 10% of the population.
       
     3. Sampling distribution approx Normal Distributions
     4. Population Standard distribution is known.
     5. Sample size>=30
     
#### Condition for accepting z value:
    p-value> Significance value
    Zcalculate< Zcritical
        
     

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm 
from statsmodels.stats import weightstats
import scipy.stats as stats #we dont have z-test in scipy library
```

A machine is producing perfume bottles with the long term average of 150cc and the standard deviation of 2 cc.
Four bottles were mpicked and the average volume was found to be 153.8cc.
Has the average volume changed from 150cc? Test with 95% confidence level.

Null Hypothesis: Mean volume = 150cc
Alternate Hypothesis: Mean volume != 150cc
Alpha = 1 - 0.95 = 0.05

Z calculated = (Observed mean - population mean) / standard error
###  z = (x  ) / ( / n)

```python
Z_cal = (153.8 - 150) / (2/np.sqrt(4))
# import math
# Z_cal = (153.8 - 150) / (2/math.sqrt(4))
Z_cal
```

```python
#norm.pdf(x, loc=0, scale=1) #Probability density function.

print(norm.cdf(x=1.64, loc=0, scale=1))  #Cumulative distribution function. (area to the left of x)

print(norm.sf(x=1.64, loc=0, scale=1)) #Survival function (also defined as 1 - cdf, but sf is sometimes more accurate).

print(norm.ppf(q=0.9494974165258963, loc=0, scale=1)) #Percent point function (inverse of cdf  percentiles).

norm.isf(q=0.9494974165258963, loc=0, scale=1) #Inverse survival function (inverse of sf).
```

# One sample t test
### Conditions for t test
    1. Random samples.
    2. Each observation should be independent of each other
        - Sampling with replacement
        - If sampling without replacement, the sample size should not be more than 10% of the population.
       
     3. Sampling distribution approx Normal Distributions
     4. Population Standard distribution is unknown.
     5. Sample size<30
     

#### Condition for accepting z value:
    p-value> Significance value
    tcalculate< tcritical
    
    dof=n-1
    
The statistic is calculated as (np.mean(a) - popmean)/se, where se is the standard error.The statistic is calculated as (np.mean(a) - popmean)/se, where se is the standard error. This is tcalculated

```python
volume = pd.Series([148.5, 153.4, 150.9, 151.2])
volume.describe()
```

```python
stats.ttest_1samp(volume, 150)
```

# One proportion test

    ### Conditions for one proportion test
    1. Random samples.
    2. Each observation should be independent of each other
        - Sampling with replacement
        - If sampling without replacement, the sample size should not be more than 10% of the population.
       
     3. Data contains only two category pass or fail
     4. The data are a simple random sample from the population of interest. The population is at least 10 times as large as the sample. np10 and n(1p)10 , where n is the sample size and p is the true population proportion.

     

```python
#Case1
#Has smoking habit changed?
#Null Hypothesis: The smoking habit has not changed
#Alternate Hypothesis: Smoking habit has changed
stats.binom_test(14,100,p=0.21,alternative='two-sided')
```

```python
#P-value in this case is 0.05
#Our p-value is 0.1092081572082592

#So we fail to reject the Null Hypothesis. It means the smoking habit has not changed
```

```python
#Case2
#Has smoking habit reduced?
#Null Hypothesis: The smoking habit has not reduced
#Alternate Hypothesis: Smoking habit has reduced
stats.binom_test(14,100,p=0.21,alternative='less')
```

```python
#P-value in this case is 0.05
#Our p-value is 0.0504246022155211

#So we fail to reject the Null Hypothesis. It means the smoking habit has not reduced
```

### F- test: we use F-test statistics for two variance test

    2. Fortesting equality of saveral mean with technique of ANOVA

### A sample of 51 bottles was selected. The standard deviation of these bottles was 2.35 cc. Has it increased from established 2cc? 90% confidence interval.

```python
x_ax=np.linspace(0,100,101)
x_ax
```

```python
y_ax=stats.chi2.pdf(x_ax,df=51) # More the df it will more towards the normal
sns.relplot(x=x_ax,y=y_ax,kind='line')
plt.show()
```

```python
y_ax=stats.chi2.pdf(x_ax,df=5) # If df reduces curve becomes steeper
sns.relplot(x=x_ax,y=y_ax,kind='line')
plt.show()
```

```python
chi_sq_cal=(51-1)*(2.35**2)/(2**2)
chi_sq_cal
```

```python
stats.chi2.isf(0.10,50)

# Since chi_sq_cal>stats.chi2.isf(0.10,50)
# we will reject null hypothesis
```

# Two-sample t test.
There are two type of t test:
1.  Two sample t test:      If the values in one sample revels no information about those of other sample, then the sample are independent.
    Ex: volume produced by two machine
    
2. Paired t test:      If the values in one sample affect the value in the other sample, then the samples are dependent.
    Ex: Blood pressure  before and after a specific medicine
    


If assumes variance of both sample is equal:


# Two Sample t test(Equal Variance)

```python
m1=[150,152,154,152,151]
m3=[144,162,177,150,140]
```

```python
stats.ttest_ind(m1,m3,equal_var=False)
```

# Two Sample t test(Unequal Variance)

```python
m1=[150,152,154,152,151]
m3=[144,162,177,150,140]
stats.ttest_ind(m1,m3,equal_var=False)
```

# Paired t test



```python
before=[120,122,143,100,109]
after=[122,120,141,109,109]
stats.ttest_rel(before,after)
```

### We fail to reject the Null hypothesis.
Ho: before = after
Ha: before  after

# Two proportion test (same conditions for each sample as 1 proportion)

There are 2 types of two proportion test:
1. yes or pooled two proportion test.
    H0: p1-p2=0
    Ha: p1-p2 !=0


2. No or un-pooled proportion test
     H0: p1-p2=d
    Ha: p1-p2 !=d


```python
from statsmodels.stats import proportion
```

```python
proportion.test_proportions_2indep(30,200,10,100)

# H0: p1-p2=0
# Ha: p1-p2 !=0
# we fail to reject null hypothesis
```

# Annova test(for more than two samples)..Analysis of variance


H0: All means are equal

Ha: Atleast one of the means is different from others

```python
from statsmodels.stats import oneway as oneway
```

```python
m1=[150,151,152,151,150]
m2=[153,152,148,151,149,152]
m3=[156,154,155,156,157,155]
np.mean(m1),np.mean(m2),np.mean(m3)
```

# Goodness of fit(chi-square)
-> To check if sample is comming from a population with specific distribution

H0: The data follows a specified distribution.

H1: The data do not follow specific distribution.





```python
# flipping of coins 100 times
exp=[50,50]
observed=[40,60]

stats.chisquare(observed,exp)
# since pvalue<0.05, coin is baised.
```

```python
## rolling a dies
obs=[10,6,8,22,11,8]
```

```python
stats.chisquare(obs) # if we give only observed value , it will fill expected as equal probability, here "1/6

# so we reject null hypothesis
```

```python
exp_ratio=pd.Series([0.1,0.2,0.4,0.3])
obs=pd.Series([25,41,91,68])
exp=exp_ratio*sum(obs)
exp
```

```python
stats.chisquare(obs,exp)
# we will except the null hypothesis.
```

# Contigency tables

-> Null hypothesis is that there is no relationship between the rows and the columns.

-> Alternate hypothesis is that there is a relationship. Alternate hypothesis does not tell what type of relationship exists.





```python
# EX:
data=np.array([[22,26,23],[28,62,26],[72,22,66]])
data
```

```python
stats.chi2_contingency(data)
# since p value is very less, there is relationship between rows and column
```
