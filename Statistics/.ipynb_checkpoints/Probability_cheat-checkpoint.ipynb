### Central Limit Theorem with a BiModal Distribution

```python
a = np.random.normal(100, 5, 1000)
b = np.random.normal(80, 5, 1000)
bm_data = np.append(a, b) # joined two array
np.random.shuffle(bm_data)
print("Mean_1000:",np.mean(bm_data ),"  Std_1000:", np.std(bm_data ))
sns.displot(bm_data)
plt.xlim(65,120)
```

```python
# Normal Distrbution  for 1000 samples with 2 datapoints
Thousand=bm_data.reshape(1000,2).mean(axis=1)
print("Mean_1000:",np.mean(Thousand),"  Std_1000:", np.std(Thousand))
print("Std_derived from population's Std: ", np.std(bm_data )/np.sqrt(2))
sns.histplot(x=Thousand,kde=True)
plt.xlim(60,120,5)
```

```python
## as sample size increase std reduces and normal distribution curves shrinks.
Two_50=bm_data.reshape(250,8).mean(axis=1)
print("Mean_1000:",np.mean(Two_50),"  Std_1000:", np.std(Two_50))
print("Std_derived from population's Std: ", np.std(bm_data )/np.sqrt(8))
sns.histplot(x=Two_50,kde=True)
plt.xlim(60,120,5)
```

### Binomial Distribution

    binomial(n, p, size=None)

    Draw samples from a binomial distribution.

    Samples are drawn from a binomial distribution with specified
    parameters, n trials and p probability of success where
    n an integer >= 0 and p is in the interval [0,1]. (n may be
    input as a float, but it is truncated to an integer in use)
    size is no of times trial has been done
    Gives from n times how many times true occurs

```python
np.random.binomial(1,0.5)
```

### Now we will use the theoritical formula of poisson

```python
from scipy.stats import poisson
#poisson.cdf(k,mu) #cumulative distribution function- for less than or equal to
#poisson.pmf(k,mu) #probability mass function- for exact value
# poisson.sf(k,mu) # Survival function- (1-cdf)
print(poisson.mean(20,0.12))
print(poisson.var(20,0.12))
print(poisson.std(20,0.12))
```

```python
#poisson.pmf(k,mu) where k = how many people we want to check in the queue and mu is the mean 
poisson.pmf(7,3.6) # probability of getting 7 people in the queue
```

```python
poisson.cdf(7,3.6) # probability of getting <=7 people in the queue
```

```python
print(poisson.sf(7,3.6)) # probability of getting more than 7 people in the queue 
# cdf and sf sum up to 1
print(poisson.sf(7,3.6)+poisson.cdf(7,3.6))
poisson.var(3.6)
```

## Questions: Perfume bottles are filled with an average volume of 150cc and a standard deviation of 2cc. 
What percent of bottles will have a volume of more than 153 CC?


```python
from scipy.stats import norm
norm.sf(153,150,2) # x=153, u=150, sd=2 prob og getting >153 cc
```

```python
#What percent of models will have the volume between 148 and 152 CC?
norm.sf(148,150,2) - norm.sf(152,150,2)# Area to the right
```

```python
#norm.cdf(x,n,p) #cumulative distribution function- for less than or equal to 2
#norm.pdf(x,n,p) #probability density function- for exact values
#norm.sf(0,20,0.12) # Survival function- more than 2 
#binom.mean(20,0.12)
# binom.var(20,0.12)
#binom.std(20,0.12)
```

```python
norm.pdf(153,150,2)  #This gives you the y axis value which is f(x)
```

```python
x_ax=np.linspace(144,156,13) # creating 13 items b/w 144 to 156
x_ax
```

```python
y_ax=norm.pdf(x_ax,150,2)  #This gives you the z table value (area on the right)
y_ax
```

```python
sns.lineplot(x=x_ax,y=y_ax)
plt.show()
```
