#### 11.Stack and concate

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
# axis = 0 - rowise - np.vstack
# axis = 1 - columnwise - np.hstack
pd.concat( (india_weather, us_weather), axis=0 )
```

```python
pd.concat((india_weather, us_weather), axis=0, ignore_index=True)
```

```python
pd.concat((india_weather, us_weather), axis=1)
```

```python
# 2 level indexing
weather_compare=pd.concat((india_weather, us_weather), axis=0, keys=['india','us'])
weather_compare
```

```python
print(weather_compare["city"]["us"][1],"\n""\n")
print(weather_compare["city"]["us"])# city chahiye us ka
print(weather_compare["city"])
# print(weather_compare["us"])## error us ka chahiye par kya
```
